```python
import pandas as pd
```

```python
def research_agent(topic):
    
    research_data = {
        "topic": topic,
        "key_points": [
            f"Overview of {topic}",
            f"Benefits and challenges of {topic}",
            f"Common questions about {topic}"
        ],
        "questions": [
            f"What is {topic}?",
            f"Why is {topic} important?",
            f"How can beginners start with {topic}?"
        ]
    }
    
    return research_data
```

```python
def brief_agent(topic, research):
    
    brief = {
        "topic": topic,
        "target_audience": "Beginners and interested readers",
        "content_goal": f"Provide clear understanding of {topic}",
        "sections": research["key_points"]
    }
    
    return brief
```

```python
def outline_agent(topic, brief):
    
    outline = [
        f"H1: {topic}",
        "H2: Introduction",
        "H2: Key Concepts",
        "H2: Benefits and Use Cases",
        "H2: Practical Tips",
        "H2: Frequently Asked Questions"
    ]
    
    return outline
```

```python
topic = "Best Budget Hotels in Mumbai"

research = research_agent(topic)

brief = brief_agent(topic, research)

outline = outline_agent(topic, brief)

print("RESEARCH OUTPUT\n")
print(research)

print("\nCONTENT BRIEF\n")
print(brief)

print("\nPAGE OUTLINE\n")
for item in outline:
    print(item)
```
